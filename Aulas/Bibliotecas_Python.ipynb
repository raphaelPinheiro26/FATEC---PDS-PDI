```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
A = 2
B = 0.5
C = -1
D = 3

# Função Exponencial
def exponential_function(x):
    return A * np.exp(B * x + C) + D

# Dados
x = np.linspace(-5, 5, 400)
y = exponential_function(x)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$f(x) = A \cdot e^{(Bx + C)} + D$', color='blue')
plt.title('Função Exponencial')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
A = 3
B = 10
C = 2
D = 5

# Função Logarítmica
def logarithmic_function(x):
    return A * np.log10(x - C) + D

# Dados
x = np.linspace(C + 0.1, 10, 400)  # Evitar logaritmo de números não positivos
y = logarithmic_function(x)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$f(x) = A \cdot \log_{B}(x - C) + D$', color='green')
plt.title('Função Logarítmica')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
A = 4
B = 2
C = np.pi / 4
D = 1

# Função Seno
def sine_function(x):
    return A * np.sin(B * (x - C)) + D

# Dados
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y = sine_function(x)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$f(x) = A \cdot \sin(B(x - C)) + D$', color='red')
plt.title('Função Seno')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
A = 5
B = 0.5
C = -np.pi
D = -3

# Função Cosseno
def cosine_function(x):
    return A * np.cos(B * (x - C)) + D

# Dados
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y = cosine_function(x)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$f(x) = A \cdot \cos(B(x - C)) + D$', color='purple')
plt.title('Função Cosseno')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## Dica:  
Você pode fazer barras de roalgem para alterar o valor em tempo real de cada parâmetro

```python
# Importa a biblioteca necessária para criação de sliders e widgets
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual

# Função de plotagem
def plot_function(A=1, B=1, C=0, D=0):
    # Cria um array de valores de x que vão de -10 a 10
    x = np.linspace(-10, 10, 400)
    # Calcula os valores de y usando a função exponencial com os parâmetros fornecidos
    y = A * np.exp(B * x + C) + D

    # Cria uma nova figura para o gráfico
    plt.figure(figsize=(10, 6))
    # Plota x versus y no gráfico, com uma legenda que inclui os parâmetros
    plt.plot(x, y, label=f'$f(x) = {A} \cdot e^{{({B}x + {C})}} + {D}$')
    # Define o título do gráfico
    plt.title('Função Exponencial')
    # Define o rótulo do eixo x
    plt.xlabel('x')
    # Define o rótulo do eixo y
    plt.ylabel('f(x)')
    # Adiciona uma grade ao gráfico para melhor visualização
    plt.grid(True)
    # Adiciona a legenda ao gráfico
    plt.legend()
    # Mostra o gráfico
    plt.show()

# Definição dos widgets para sliders
# Slider para o parâmetro A, com valor inicial 1, mínimo 0, máximo 10 e passo de 0.1
A_slider = widgets.FloatSlider(value=1, min=0, max=10, step=0.1, description='A:')
# Slider para o parâmetro B, com valor inicial 1, mínimo -5, máximo 5 e passo de 0.1
B_slider = widgets.FloatSlider(value=1, min=-5, max=5, step=0.1, description='B:')
# Slider para o parâmetro C, com valor inicial 0, mínimo -5, máximo 5 e passo de 0.1
C_slider = widgets.FloatSlider(value=0, min=-5, max=5, step=0.1, description='C:')
# Slider para o parâmetro D, com valor inicial 0, mínimo -10, máximo 10 e passo de 0.1
D_slider = widgets.FloatSlider(value=0, min=-10, max=10, step=0.1, description='D:')

# Cria uma interface interativa que conecta os sliders à função de plotagem
interactive_plot = interactive(plot_function, A=A_slider, B=B_slider, C=C_slider, D=D_slider)

# Obtém o widget de saída da interface interativa
output = interactive_plot.children[-1]
# Ajusta a altura da área de saída para melhorar a visualização do gráfico
output.layout.height = '500px'
# Mostra a interface interativa com os sliders
interactive_plot
```
