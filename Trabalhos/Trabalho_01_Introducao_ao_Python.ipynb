## 1 - Sistema de Senhas para Pronto Atendimento

### Introdução

Desenvolva um sistema avançado para gerenciar a distribuição de senhas no Atendimento de Emergência de um hospital. O sistema deve não apenas distribuir senhas para diferentes filas, mas também gerenciar prioridades, gerar relatórios detalhados e oferecer funcionalidades adicionais para melhorar a eficiência e a experiência do atendimento.

### Funcionalidades Básicas:

1. **Filas de Atendimento**:
   * O hospital possui 3 filas distintas:
     * **Ortopedia**
     * **Atendimento Comum**
     * **Atendimento Prioritário**
   * Senhas no formato `AAA-000`, onde `AAA` representa a fila e `000` representa a posição na fila:
     * **Atendimento Prioritário**: `ATP`
     * **Atendimento Comum**: `ATC`
     * **Ortopedia**: `ORT`

2. **Geração e Exibição de Senhas**:
   * Gerar senhas e adicionar à fila correspondente.
   * Exibir a senha gerada na tela com a posição atual na fila.
   * Registrar o horário de emissão da senha para controle de tempo de espera.

3. **Consulta de Filas**:
   * Permitir a visualização das senhas em cada fila, ordenadas por horário de chegada.
   * Permitir a pesquisa por senha específica, exibindo todas as informações associadas.

4. **Chamada de Pacientes**:
   * Chamar a próxima senha na fila e removê-la.
   * Registrar a data e hora da chamada.
   * Permitir a reatribuição da senha se o paciente não comparecer após um intervalo definido.

### Funcionalidades Avançadas:

5. **Prioridade Dinâmica e Escalonamento**:
   * Implementar um sistema de priorização que permita mover senhas entre filas conforme a gravidade da situação.
   * Adicionar um mecanismo para escalonamento automático baseado no tempo de espera e urgência.

6. **Notificações e Alertas**:
   * Implementar alertas automáticos para avisar os pacientes sobre a chamada da senha e espera.
   * Adicionar notificações para a equipe médica sobre filas congestionadas ou senhas antigas.

7. **Gerenciamento de Senhas**:
   * Adicionar funcionalidades para cancelar senhas e reatribuir para filas.
   * Implementar um histórico de chamadas de senhas para monitoramento e análise de desempenho.

8. **Relatórios e Análise**:
   * Gerar relatórios detalhados sobre o tempo de espera, número de pacientes atendidos e desempenho das filas.
   * Analisar padrões de atendimento e gerar insights para otimização de processos.

### Funcionalidades Opcionais:

9. **Interface Gráfica**:
   * Desenvolver uma interface gráfica que facilite a visualização das filas e gerenciamento de senhas com gráficos e filtros interativos.

10. **Alerta**
    * Ao ultrapassar o tempo médio de espera estipulado em cada fila, enviar um e-mail para o supervisor de atendimento.

11. **Integração com Dispositivos Móveis**:
   * Integrar o sistema com aplicativos móveis para envio de notificações e gerenciamento remoto das medicações.

### Introdução

Desenvolva um sistema para simular o monitoramento de sinais vitais de pacientes em um ambiente hospitalar. O programa deve permitir a entrada e registro dos sinais vitais, gerar alertas baseados em critérios estabelecidos e fornecer visualizações das variações dos sinais vitais ao longo do tempo. O sistema deve também permitir o acompanhamento histórico dos dados registrados para análise e tomada de decisão.

### Funcionalidades Básicas:

1. **Menu Principal**:
   * O programa deve apresentar um menu principal com as seguintes opções:
     * Entrada de Dados
     * Monitoramento
     * Visualização de Dados

2. **Entrada de Dados**:
   * Permitir o cadastro de múltiplos pacientes com os seguintes dados:
     * Nome do Paciente
     * ID do Paciente
   * Registrar os seguintes sinais vitais para cada paciente:
     * Frequência Cardíaca
     * Pressão Arterial
     * Saturação de Oxigênio (SpO2)
     * Temperatura Corporal
   * Armazenar a hora em que os sinais vitais foram registrados.

3. **Monitoramento**:
   * Gerar alertas se:

     * A Temperatura Corporal estiver fora da faixa de 36.1°C a 37.2°C.

   * Permitir a visualização dos sinais vitais registrados ao longo do tempo para cada paciente.

### Funcionalidades Avançadas:

5. **Gestão de Alertas Personalizados**:

     * Configurar níveis de prioridade para diferentes tipos de alertas.

6. **Histórico e Auditoria**:
   * Manter um histórico detalhado de todos os registros de sinais vitais e alertas gerados.

     * Sinais vitais registrados e alertas gerados.
     * Tendências e variações ao longo do tempo.
     * Análise dos dados para avaliar a eficácia do monitoramento e a resposta aos alertas.

### Funcionalidades Opcionais:

8. **Interface Gráfica Avançada**:
   * Desenvolver uma interface gráfica intuitiva para facilitar a entrada de dados, o monitoramento e a visualização dos sinais vitais.

## 6 - Sistema de Gestão de Certificados de Equipamentos Biomédicos

### Introdução

Desenvolva um programa para gerenciar os certificados de calibração e testes de segurança dos equipamentos médicos em um hospital. O sistema deve permitir a administração detalhada das informações relacionadas à validade dos certificados, incluindo requisitos específicos para equipamentos radiológicos e da Central de Material Esterilizado (CME). O programa também deve oferecer funcionalidades para consulta, alerta e geração de relatórios sobre a validade dos certificados.

### Funcionalidades Básicas:

1. **Menu Principal**:
   * O programa deve apresentar um menu principal com as seguintes opções:
     * Cadastro de Equipamentos
     * Registro de Certificados
     * Consulta de Equipamentos
     * Relatórios

2. **Cadastro de Equipamentos**:
   * Inserir as seguintes informações para cada equipamento:
     * Nome do Equipamento
     * TAG de Identificação
     * Tipo de Equipamento (Radiológico, CME, ou Outro)
     * Local de Uso
     * Data de Aquisição
   * Exemplo de TAG: UTI-MNP-0003.

3. **Registro de Certificados**:
   * Registrar os seguintes certificados para cada equipamento:
     * **Certificado de Calibração**: Data de Emissão e Data de Validade.
     * **Teste de Segurança Elétrica**: Data de Emissão e Data de Validade.
     * Para equipamentos radiológicos, adicionar:
       * **Laudo Radiológico**: Data de Emissão e Data de Validade.
       * **Inspeção Radiométrica**: Data de Emissão e Data de Validade.
     * Para equipamentos da CME, adicionar:
       * **Qualificação Térmica**: Data de Emissão e Data de Validade.

4. **Consulta de Equipamentos**:
   * Permitir busca de equipamentos por:
     * TAG
     * Tipo de Equipamento
   * Exibir:
     * Todas as informações cadastrais do equipamento.
     * Status e validade dos certificados relacionados (Calibração, Teste de Segurança, Laudo Radiológico, Inspeção Radiométrica, Qualificação Térmica).

### Funcionalidades Avançadas:

5. **Alertas de Expiração de Certificados**:
   * Implementar alertas automáticos quando a validade de qualquer certificado estiver prestes a expirar (por exemplo, 30 dias antes da expiração).
   * Enviar notificações via e-mail ou mensagens para os responsáveis.

6. **Gestão de Manutenções e Indicadores**:
   * Adicionar funcionalidades para registrar e acompanhar as manutenções realizadas nos equipamentos.
   * Incluir indicadores de desempenho, como MTBF (Tempo Médio Entre Falhas) e MTTR (Tempo Médio para Reparo).

7. **Histórico e Auditoria**:
   * Manter um histórico detalhado de todos os registros de certificados e manutenções.
   * Implementar um sistema de logs para registrar todas as alterações e acessos ao sistema.

8. **Relatórios Avançados**:
   * Gerar relatórios detalhados sobre:
     * Equipamentos com certificados prestes a expirar.
     * Histórico de certificados e manutenções.
     * Indicadores de desempenho dos equipamentos.

### Funcionalidades Opcionais:

9. **Interface Gráfica Avançada**:
   * Desenvolver uma interface gráfica intuitiva para facilitar o gerenciamento dos certificados e a visualização dos dados.

10. **Integração com Dispositivos Móveis**:
   * Integrar o sistema com aplicativos móveis para envio de notificações e gerenciamento remoto.
